```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = 6 * np.random.rand(100, 1) - 3
y =  0.5*X**2 + 1.5*X + 2 + np.random.rand(100, 1)
```

```python
plt.scatter(X, y, color='r')
plt.xlabel("X dataset")
plt.ylabel("y dataset")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,
                                                    random_state=40)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg1 = LinearRegression()
```

```python
linreg1.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = linreg1.predict(X_test)
score = r2_score(y_test, y_pred)
print(score)
```

```python
# visualize the model
plt.plot(X_train, linreg1.predict(X_train), color='b')
plt.scatter(X, y, color='r')
plt.xlabel("X dataset")
plt.ylabel("y dataset")
plt.show()
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias = True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)
```

- The fit_transform method is used on the training data to fit the preprocessing technique and transform the data, learning from the training dataset's characteristics.
- The transform method is applied on the test data using the parameters learned from the training data's fit_transform, ensuring that the same transformations are applied to maintain consistency and avoid information leakage between training and test sets.

#### Polynomial Regression and Feature Engineering:
- Polynomial regression involves creating polynomial combinations of existing features to capture nonlinear relationships between the predictors (input features) and the target variable.
- For instance, if you have a single feature 'x', polynomial regression might create additional features like 'x^2', 'x^3', etc., to model curved relationships between 'x' and the target.

```python
X_train_poly
```

```python
linreg2 = LinearRegression()
linreg2.fit(X_train_poly, y_train)
y_pred = linreg2.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```

```python
linreg2.coef_
```

```python
# visualize the model
plt.scatter(X_train, linreg2.predict(X_train_poly), color='b')
plt.scatter(X_train, y_train, color='r')
plt.xlabel("X dataset")
plt.ylabel("y dataset")
plt.show()
```
